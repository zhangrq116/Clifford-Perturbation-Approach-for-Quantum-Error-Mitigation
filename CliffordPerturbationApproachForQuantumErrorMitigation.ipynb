```python
# Fig4 in main text

# Fig4a in main text



with open('main_fig4a.pkl', 'rb') as f:
    x_angles        = pickle.load(f)
    y_unmit_exp_list= pickle.load(f)
    y_pred          = pickle.load(f)
    y_mit_exp       = pickle.load(f)
    train_y         = pickle.load(f)
    MPS_sim         = pickle.load(f)
    isoTNS_sim      = pickle.load(f)
    exact_sim       = pickle.load(f)
    pauli           = pickle.load(f)



fontsize_global=24
plt.rcParams['font.size'] = fontsize_global
plt.rcParams['axes.linewidth'] = 2

ylim = [-0.1, 1.15]
xfactor = 1
alpha=1.0
markersize_factor=4.0
markersize=2*markersize_factor
markersize_sim=100*markersize_factor-300
markersize_ideal=1000*markersize_factor
linewidth=6


pauli_string=[[], [], []]
pauli_string[0]=[str(i) for i, p in enumerate(pauli) if p.upper() == 'X']
pauli_string[1]=[str(i) for i, p in enumerate(pauli) if p.upper() == 'Y']
pauli_string[2]=[str(i) for i, p in enumerate(pauli) if p.upper() == 'Z']
pauli_string_=''
for idxs, p in zip(pauli_string, ['X', 'Y', 'Z']):
    if len(idxs) > 0:
        pauli_string_+=p
        pauli_string_+='_{'
        pauli_string_+=','.join(idxs)
        pauli_string_+='}'
ylabel = '$<'+pauli_string_+'>$'

fit_colors=list(cm.Reds(np.linspace(0.3, 0.9, 3))) + ['b'] + ['g']

ncol=1; nrow=1;
fig, axes = plt.subplots(nrow,ncol, figsize=(8*ncol,8*nrow))

kwargs_c1_exp={'color': 'g', 'label': 'unmitigated', 's': markersize_sim, 'facecolors': 'none', 'linewidths': 5.0}
kwargs_exp={'color': 'b', 'label': 'mitigated by IBM', 's': markersize_sim}

ax = axes

ax.axvline(np.pi/2, color='grey', linestyle='--', linewidth=0.5)
ax.axvline(0, color='grey', linestyle='--', linewidth=0.5)
ax.axhline(1, color='grey', linestyle='-', linewidth=0.5)
ax.axhline(0, color='grey', linestyle='-', linewidth=0.5)

# tensor network method
x, y = MPS_sim
marker='s'; color='magenta'; edgecolor='none'
linewidth_=linewidth
label_='MPS'
idx = np.argsort(x)
x = np.array(x)[idx]
y = np.array(y)[idx]

x, y = isoTNS_sim
marker='^'; color='tomato'; edgecolor='none'
linewidth_=linewidth-3
label_='isoTNS'
idx = np.argsort(x)
x = np.array(x)[idx]
y = np.array(y)[idx]

# exact simulation results
x, y = exact_sim
linewidth_=linewidth+3
ax.plot(x, y, color='dimgrey', marker='.', label='exact', linewidth=linewidth_-2.5, zorder=1)



kwargs_c1_exp_mini={'color': 'r', 'label': 'mitigated by our CPDR', 's': markersize_sim, 'facecolors': 'none', 'linewidths': 5.0,'marker': '^'}


cmap = plt.cm.viridis  
num_plots = len(y_unmit_exp_list)  
colors = cmap(np.linspace(0, 1, num_plots))  
noise_factor_list = [1,1.2,1.6]
for i, y_data in enumerate(y_unmit_exp_list):
    if i > 0:
        break
    ax.scatter(x_angles, y_data, color=colors[i+1], label = 'unmitigated, G ='+str(noise_factor_list[i]), s= markersize_sim, linewidths = 5.0,zorder=3 - i)





ax.scatter(x_angles, y_pred, **kwargs_c1_exp_mini, zorder=10)

ax.scatter(x_angles, y_mit_exp, **kwargs_exp, zorder=4)
                
ax.set_xlim(np.array([np.min(x_angles)-0.05, np.max(x_angles) + 0.05])*xfactor)
ax.set_ylim(ylim)  

xticks = np.array([0, 0.125, 0.25, 0.375, 0.5])*np.pi
xticks_label = ['0', '$\pi$/8', '$\pi$/4', '$3\pi$/8', '$\pi$/2']
ax.tick_params(axis="both", direction="in", length=8, width=2, pad=15)
ax.set_xticks(xticks);ax.set_xticklabels(xticks_label)
ax.set_xlabel('Rx angle $\\theta_h$')




ax.set_ylabel(r'$\left\langle X_{\{3\}} Y_{\{2\}} Z_{\{5\}}\right\rangle$',fontsize=fontsize_global+6) 

from mpl_toolkits.axes_grid1.inset_locator import inset_axes
ax_inset = inset_axes(ax, width="70%", height="70%", bbox_to_anchor=[117, 203, 300, 300])  

ax_inset.plot(x_angles, np.abs(y_mit_exp - train_y), color='b',marker = '.')  
ax_inset.plot(x_angles, np.abs(y_pred - train_y), color='r',marker = '^')  
ax_inset.set_yscale('log')


xticks = np.array([0, 0.125, 0.25, 0.375, 0.5])*np.pi
xticks_label = ['0', '$\pi$/8', '$\pi$/4', '$3\pi$/8', '$\pi$/2']
ax_inset.tick_params(axis="both", direction="in", length=8, width=2, pad=15)
ax_inset.set_xticks(xticks);ax_inset.set_xticklabels(xticks_label)



```
